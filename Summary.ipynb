# Data Processing

## Download Dataset (downloadDataset.py)

- url: http://data.caida.org/datasets/topology/ark/ipv4/probe-data/
- Iterate through all cycles and download the data on the first day of each month. For 2014.09 and 2016.07, since there is no data on 1st, dataset move to the first existing day: 2014/09/12 and 2016/07/02.

## Rawdata to Dataframe (EditDatabaseFirstDayEachMonth.ipynb)

- "F:/prob/": Raw data 
- "F:/dataset/": Upzip ".gz"
- "F:/dataset_YYYYMMDD/": For each record in .warts, it has an variable **src_address** and another variable **hops**(with attributes **address** and **rtt**). Save data in a pandas.dataframe with columns **IP_SRC**, **IP_DST**, and **RTT**.
- "F:/dictionary_ip.pkl": dictionary of IP.

Note: 
1. to_df_for_t3() is much faster than to_df().
2. 2014/09/12 and 2016/07/02 added.

## Modift Dataset (ModifyDB.ipynb)

1.For each month, we collect all data for the first existing day.
2.Pick the top 20% nodes that appear frequently
3.Keep the edges that both nodes are in the top 20% node set.
4.Let the weight of each edge to be the median of the corresponding RTTs.

### Subset: Top 20% Frequent Nodes
- whole.pkl(67069411) -> top20.pkl(13413882): go through all data and find the top 20 frequent nodes.
- "F:/dataset_YYYYMMDD_sub/": keep only the rows that contains both nodes in provided node dataset.

### Subset: Median of RTT
- "F:/dataset_YYYYMMDD_median_undirected/": YYYYMM.pkl, order rows for each month by IP_SRC and IP_DST and keep the median of RTT for each group.

### Dataset to Graph
- "F:/graph_YYYYMMDD_median_undirected/": YYYYMM.gexf
